```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("pd_titanic_house.db") 

df1 = pd.read_csv("preprocessed_dataclassification.csv")
df2 = pd.read_csv("preprocessed_dataregression.csv")

df1.to_sql('pd_titanic', conn, if_exists='replace', index=False)
df2.to_sql('pd_house', conn, if_exists='replace', index=False)

result_1 = pd.read_sql_query("SELECT * FROM pd_titanic", conn)
display(result_1)

result_2 = pd.read_sql_query("SELECT * FROM pd_house", conn)
display(result_2)

conn.commit()
conn.close()
```

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect("pd_titanic_house.db")

# Read the preprocessed data files
df1 = pd.read_csv("preprocessed_dataclassification.csv")
df2 = pd.read_csv("preprocessed_dataregression.csv")

# Create and populate the pd_titanic table
df1.to_sql('pd_titanic', conn, if_exists='replace', index=False)

# Create and populate the pd_house table
df2.to_sql('pd_house', conn, if_exists='replace', index=False)

# Commit changes and close the connection
conn.commit()
conn.close()
```
